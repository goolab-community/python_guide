# დოკუმენტაციის პრინციპები
მიუხედავად იმისა რომ პითონი ერთ-ერთ კარგად წაკითხვადი პროგრამირების ენაა 
(თითქმის ფსევდოკოდის ექვივალენტურია), მაინც დროთა განმავლობაში ჩვენ თვითონვე 
გვავიწყდება თუ რა ფუნქცია აკისრია ან როგორ მუშაობს კოდის გარკვეული ნაწილი. 
ასევე გაცილებით უფრო დიდ პრობლემა წარმოიქნება როდესაც სხვას უწევს მოცემული 
კოდის გარჩევა და გამოყენება. 
აქედან გამომდინარე მიზანშეწონილია და ზოგ შემთხვევაში აუცილებელიც, რომ პროგრამულ 
კოდს მოვაყოლოთ განმარტებებიც ე.წ. "დოქსტრინგის" სახით, რომელიც მიღებულია რომ 
დაიწეროს მრავალ-ხაზიანი სტრინგის ჩანაწერით, სამმაგ ბრჭყალებში მოქცეული ტექსტით.

რა თქმა უნდა განმარტებების წერა შეგვიძლია ისე როგორც ჩვენ მოგვესურვება კომენტარების 
სახით, მაგრამ არსებობს დოქსტრინგის წერის პრინციპები და სტანდარტები, რომლის გათვალიწინებაც 
კოდს გაცილებით წაკითხვადს და მარტივად გასაგებს გახდის.

პირველ რიგში როდესაც კოდის დოქსტრინგზე ვსაუბრობთ, აუცილებლად უნდა გავმიჯნოთ ის 
სტანდარტული ერთ ხაზიანი კომენტარისაგან, რომელიც იწყება # სიმბოლოთი. 

აუცილებელია დავიმახსოვროთ რომ კომენტარი ჩანს მხოლოდ მაშინ როდესაც კოდი გახსნილი 
გვაქვს რომელიმე ედიტორით ან IDE-ით, ხოლო რანთაიმში კოდის მუშაობის დროს ან 
კომპილაციის შემდგომ ის აღარ გამოჩნდება. 

როგორც ვიცით იმისთვის რომ გავიგოთ თუ როგორ მუშაობს რომელიმე მოდული, ბიბლიოთეკა, 
კლასი ან თუნდაც მეთოდი, ვიყენებთ help ფუნქციას, რომელსაც დოკუმენტაციის ტექსტი მოაქვს
დოქსტრინგიდან (და არა კომენტარებიდან), მაგალითად: 

მრავალხაზიან კომენტარში მოქცეული სტრინგი გამოჩნდება help(TestHello)
ფუნქციით როგორც დოქსტრინგი. 

```python
class TestHello:
    def __init__(self):
        """ This is a dostring comment"""
        print('Hello, world!')
```

კომენტარის მაგალითი რომელიც არ გამოჩნდება help(TestHello)

```python
class TestHello:
    def __init__(self, a, b):
        # This is just a comment text
        print('Hello, world!')
```

დოქსტრინგის წერის დროს აუცილებელია დავიმახსოვროთ შემდეგი პრინციპები: 

* დოქსტრინგი არ უნდა იმეორებდეს ფუნქციის სტრუქტურას (გადასაცემ პარამეტრებს, 
ფუნქციის დასახელებას და ა.შ.) რაც ისედაც კარგად ჩანს. 
* დოქსტრინგში მოცემული ინფორმაცია, აუცილებლად უნდა იყოს კონკრეტულად და 
ნათლად მოცემული.

help ფუნქციის გარეშე დოქსტრინგს შეგვიძლია მივწვდეთ ობიექტის __doc__ ატრიბუტიდან.

#### დოქსტრინგის წერის პოპულარული სტანდარტებია: 
* Epytext
* reST
* Google docstring
* Numpydoc

### Numpydoc დოქსტრინგ სტანდარტი

numpydoc არის ერთ-ერთი ყველაზე მეტად გავრცელებული სტანდარტი. სხვა სტანდარტებისგან განსხვავებით numpydoc სტანდარტით დოკუმენტირებული კოდი კარგად გამოიყურება როგორც ტერმინალში ასევე უბრალო ტექსტ ედიტორებშიც.
მოცემული სტანდარტის მიხედვით დოქსტრინგის სექციები გამოყოფილია ერთმანეთისგან სათაურებით (გარდა ე.წ. Deprecation worning-ისა). ყოველი სათაური გამოყოფილია "-" სიმბოლოებით, ხოლო სექციებისა და სათაურების თანმიმდევრობა უნდა იყოს შემდეგნაირი: 

1. #### Short summary - მოკლე აღწერა
 ერთ-ხაზიანი აღწერა რომელიც არ შეცავს არც ობიექტის (შესაბამისი კლასის, ფუნქციის ან მოდულის) და პარამეტრების სახელებს, მაგალითად: 
 
``` python
def add(a, b):
    """The sum of the numbers.
    
    """
```

ფუნქციის სიგნატურა (ფუნციის დასახელება და გადასაცემი პარამეტრები), როგორც წესი ჩანს ინტროსპეციით, მაგალითად help ფუნქციის გამოყენებით. ამიტომ მოკლე აღწერაში მისი ხსენება არ არის საჭირო. გამონაკლისია ის შემთხვევა როდესაც მოდული დაწერილია C ენაზე, ასეთ დროს აუცილებელია მოკლე აღწერის ტექსტის პირველივე ხაზში მისი მითითება, მაგალითად: 

``` python
"""
add(a, b)

The sum of the numbers.

"""
```

2. #### Deprecation warning 
Deprecation (მოძველებული) warning - სექცია გვატყობინებს რომ ობიექტი მოძველებულია და არ არის სასურველი მისი გამოყენება (ობიექტების მსგავსად მონიშვნა გამოიყენება მაშინ როდესაც კოდის შემდგომ ვერსიაში დაგეგმილია მისი ჩანაცვლება ან წაშლა). სექცია მოიცავს: 

 - რომელ ვერსიაში მიენიჭა ობიექტს სტატუსი მოძველებული
 - მოძველების მიზეზი
 - გზა თუ რით შეიძლება ჩავანაცლოდ იგივე ობიექტი ან ფუნცია მომავალში. 

სექციის სათაური აღინიშნება deprecated დირექტივით, შემდეგნაირად:

```
.. deprecated:: 1.6.0
          `ndobj_old` will be removed in <your_module_name> 2.0.0, it is replaced by
          `ndobj_new` because the latter works also with array subclasses.
```

3. #### Extended Summary - ვრცელი მიმოხილვა
მოცემულ სექციაში, განვიხილავთ ობიექტის ფუნქციონალს რაც შეიძლება კონკრეტულად, ხოლო ვრცელი აღწერისთვის გამოყოფილია ცალკე Notes სექცია. ასევე შესაძლებელია ფუნქციის ან ობიექტის სახელის და პარამეტრების მითითება, მაგრამ აუცილებელია ვიცოდეთ, რომ პარამეტრების აღწერისთვის გამოყოფილია ცალკე Parameters სექცია.

4. #### Parameters - პარამეტრები 
მოცემულია ფუნქციის ან კონსტრუქტორის არგუმენტების აღწერა, ქეივორდები (keyword) და მათი შესაბამისი ტიპები. მაგალითად: 

```

----------

    Description of parameter `x`

    Description of parameter `y` (with type not specified).
```
აღწერაში ცვლადის სახელი მოქცეულია \` ნიშნის ქვეშ. მეორე შემთხვევაში y ცვლადი აღწერილია ტიპის მითითების გარეშე.
პარამეტრის ტიპი აუცილებელია რომ მითითებული იყოს რაც შეიძლება ზუსტად, მაგალითად ქვემოთ მოცემულია მაგალითი ყველაზე ხშირად გამოყენებული პარამეტრების ტიპებით და სავარაუდო ცვლადის სახელებით: 

``` 
Parameters
----------
filename : str
copy : bool
dtype : data-type
iterable : iterable object
shape : int or tuple of int 
files : list of str
```

თუ ქეივორდ არგუმენტის მითითება არ არის აუცილებელი, მიეთითება optional, მაგალითად: 

```
x : int, optional 
```

ასეთი პარამეტრები გამოცხადებულია საწყისი (დეფაულთ) მნიშვნელობებით, რომელიც თავისთავად აისახება ფუნქციის სიგნატურაში, მიუხედავად ამისა, შესაძლებელია მათი პარამეტრების აღწერაში მოხსენიება, მაგალითად: 

```
Description of parameter `x` (the default is -1, which implies summation
over all axes).
```

თუ პარამეტრს აქვს მხოლოდ რამოდენიმე ფიქსირებული მნიშვნელობა, ასეთი მნიშვნელობები შესაძლებელია ჩაიწეროს სისტემური ბრჭყალების შიგნით, სადაც დეფაულთ მნიშვნელობა მითითებულია პირველი, მაგალითად: 

```

```

თუ რამოდენიმე პარამეტრს აქვს ერთი და იგივე ტიპი, ფორმა და პრინციპულადაც მსგავსია, ასეთ დროს შესაძლებელია მათი გაერთიანება, მაგალითად:

```
    Input arrays, description of `x1`, `x2`.
    
```
