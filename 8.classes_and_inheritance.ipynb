### ობიექტის ატრიბუტები და მეთოდები

შევქმნათ CarFactory კლასი ხელახლა. ზემოთ მოცემული მაგალითისგან განსხვავებით, კლასში განვსაღვროთ ობიექტის ატრიბუტები.

კლასში მოცემული init ფუნქცია სრულდება ავტომატურად ობიექტის შექმნის მომენტში და ქმნის ობიექტის ატრიბუტებსაც, ხოლო ატრიბუტები ობიექტთან დაკავშირებულია self ცვლადით. როგორც ქვემოთ მოცემულ სურათზეა ნაჩვენები self ცვლადი მოუთითებს, თვით კლასის ობიექტზე.

კლასში შესაძლებელია ობიექტის მეთოდების გამოცხადება, ხოლო **იმისთვის რომ კლასის მეთოდი დაკავშირებული იყო კლასის ობიექტთან, მეთოდს პირველ პარამეტრად უნდა გადავცეთ იგივე self ცვლადი, რომლიც გადაეცემა init მეთოდს.**
ამ შემთხვევაში self არ წარმოდგენს ქივორდს და მხოლოდ სატანდარტული პირობითი დასახელებაა. self-ის მაგივრად შეგვიძლია init მეთოდის პირველ ცვლადს ვუწოდოთ ის რაც მოგვესურვება, მაგრამ არ უნდა დაგვავიწყდეს ობიექტის მეთოდებისთვის პირველ პარამეტრად იგივე დასახელების ცვლადის გადაცემა. 
ამ ყველაფრის გათვალისწინებით, არსებულ კლასს მივცეთ სრული სახე და დავამატოთ მეთოდები: buildCar, paint, get_info.

```python
class CarFactory:
    address = None
    
    def build_factory(address):
        # კლასის ატრიბუტს კლასში ვწვდებით კლასის დასახელებიდან
        CarFactory.address = address
    
    def __init__(self, car_type):
        # ყველა ობიექტის ატრიბუტს ვწვდებით self-დან
        self.car_type = car_type
        self.engine_size = None
        self.color = None
        self.assembled = False
        print('Car assembly initiated!')
    
    def buildCar(self):
        self.assembled = True
        
    def paint(self, color):
        self.color = color
        # ობიექტის მეთოდთან ისევე როგორც ობიექტის ატრიბუტებთან,
        # წვდომა გვაქვს self-დან
        self.buildCar()
        
    def get_info(self):
        print(f'Car is assembled with type: {self.car_type}, ' + 
                f'engine: {self.engine_size}, color: {self.color}')
```

## @classmethod and @staticmethod

```python
class ClassTest:
    temp = 'hello'
    
    def instance_method(self):
        print(f'Called instance_method if {self}')
    
    @classmethod
    def class_method(cls):
        # cls პარამეტი წარმოადგენს იგივე კლასს 
        print(f'{cls.temp}Called class_method of {cls}')
        cls.temp = 'world'
    
    @staticmethod
    def static_method():
        print(f'{ClassTest.temp} Called static_method.')
        
test = ClassTest()
```

```python
test.instance_method()
ClassTest.class_method()
test.class_method()
test.static_method()
ClassTest.static_method()
```

### Build car with class method

```python
class CarFactory:
    address = None
    
    def build_factory(address):
        # კლასის ატრიბუტს კლასში ვწვდებით კლასის დასახელებიდან
        CarFactory.address = address
    
    def __init__(self, car_type):
        # ყველა ობიექტის ატრიბუტს ვწვდებით self-დან
        self.car_type = car_type
        self.engine = None
        self.color = None
        self.assembled = False
        print('Car assembly initiated!')
    
    @classmethod
    def buildCar(cls):
        cls.address = 'Tesla motors facory address'
        car = CarFactory('SUV')
        car.engine = "2x245kW"
        car.color = 'navy'
        car.assembled = True
        car.get_info()
        
    def paint(self, color):
        self.color = color
        # ობიექტის მეთოდთან ისევე როგორც ობიექტის ატრიბუტებთან,
        # წვდომა გვაქვს self-დან
        self.buildCar()
        
    def get_info(self):
        print(f'Car is assembled with type: {self.car_type},' + 
                f'engine: {self.engine}, color: {self.color}')
```

```python
# In this case buildCar class method builds car object
car = CarFactory('SUV')
car.buildCar()
CarFactory.buildCar()
```

```python
class Bookshop():
    # არასწორია, books კოლექცია ინიციალიზდება კლასის გამოცხადებისას და 
    # ყოველი ობიქტის შექმნისას books კოლექცია მიებმევა კლასის კოლექციას
    # და არა ობიექტისას
    def __init__(self, name, books=[]):
        self.name = name
        self.books = books
        
    def addbook(self, book):
        self.books.append(book)
        
    def cell_book(index):
        return self.books.pop(books[index])
```

# Inheritance (მემკვიდრეობითობა)
პითონში შესაძლებელია კლასებმა მემკვიდრეობით მიიღონ სხვა კლასის ატრიბუტები და მეთოდები, რაც საკმაოდ მოხერხებული მეთოდია სხვადასხვა აბსტრაქტული ობიექტების ერთმანეთთან დასაკავშირებლად. მაგალითად გამოვაცხადოთ კლასი Animal, რომლიც იქნება სხვა და სხვა ცხოველის "წინაპარი" კლასი. 


```python
class Animal:
    def __init__(self, limbs, eyes, is_pet):
        self.limbs = limbs
        self.eyes = eyes
        self.is_pet = is_pet
        self.size = 'big'
    
    def describe(self):
        print("This animal doesn't exists!")
        
    def eat(self):
        print('Eating')
```

```python
#ობიექტის შექმნის დროს გადაცემული პარამეტრები მოდის __init__ მოთოდის limbs, eyes, is_pet პარამეტრებში
animal = Animal(2, 2, is_pet=False)

print(f'Animal have: limbs: {animal.limbs}, eyes: {animal.eyes}, Is a pet: {animal.is_pet}')
```

მემკვიდრეობითობის პრინციპის კარგად გასაგებად შევქმნათ Dog და Spider ობიექტები, რომლებიც მემკვიდრეობითობით მიიღებენ Animal-ის ყველა ატრიბუტს და მეთოდს.

```python
class Dog(Animal):
    def __init__(self, breed, role, name):
        # მშობელ კლასთან მიმათართვა ხდება ცხადი სახით, 
        # Animal.__init__(self, 4, 2, True), 
        # სადაც self პარამეტრით ხდება ორი კლასის 
        # ატრიბუტების ერთმანეთთან დაკავშირება
        # რაც არ არის რეკომდირებული
        print(self, breed, role, name)
        Animal.__init__(self, 4, 2, True)
        self.breed = breed
        self.role = role
        self.name = name
        self.pees_on_tree = True
    
    # describe მეთოდი გადაფარავს მშობელი Animal კლასის 
    # describe მეთოდს
    def describe(self):
        s = f'''
        Animal: {type(self)},
        Breed: {self.breed},
        Role: {self.role},
        Name: {self.name},
        Pees on three: {self.pees_on_tree}
        '''
        print(s)
    
    def bark(self):
        print('Bark Bark Bark!')


class Spider(Animal):
    def __init__(self, poisonous, name):
        super().__init__(8, 5, False)
        # super მეთოდი მიმართავს მშობელ Animal კლასს
        # მოცემულ შემთხვევაში, ჩანაწერი იგივეა რაც:
        # Animal.__init__(self, 8, 5, False)
        self.poisonous = poisonous
        self.name = name

    def make_net(self):
        print('Spider made a net!')
        
    def eat_fly(self):
        print('Eating fly, nam nam nam.')
```

იმისთვის რომ Dog და Spider კლასის ობიექტებმა მემკვიდრეობით მიიღონ Animal-ის ყველა ატრიბუტი, აუცილებელია super მეთოდით init მეთოდის გამოძახება. წინააღმდეგ შემთხვევაში, init მეთოდი გადაფარავს Animal კლასის init მეთოდს და მისი ატრიბუტებიც დაიკარგება, ხოლო იმ შემთხვევაში თუ, ობიექტის შექმნისას არ გადავცემთ პარამეტრებს და init მეთოდს არ გამოვაცხადებთ კლასის შიგნით, მაშინ კლასი ავტომატურად მიიღებს მშობელი კლასის ატრიბუტებს.

იმ შემთხვევაში თუ კლასს ჰყავს რამოდენიმე წინაპარი და გვსურს რომ კლასმა მემკვიდრეობით მიიღოს მხოლოდ კონკრეტული კლასის ატრიბუტები, ამისთვის შეგვიძლია super მეთოდს ცხადად გადავცეთ პარამეტრად ის კლასი რომლისგანც გვსურს მემკვიდრეობის მიღება. მაგალითად: super(Animal_parent, self), პირველი არგუმენტია მშობლის მშობელი კლასი, ხოლო მეორე საკუთარი ობიექტი.

## მრავლობითი მემკვიდრეობის დასკვნითი მაგალითი

```python
class Animal:
    def __init__(self, limbs=None, eyes=None, is_pet=None):
        self.limbs = limbs
        self.eyes = eyes
        self.is_pet = is_pet
        self.size = 'big'
    
    def describe(self):
        print("This animal doesn't exists!")
        
    def eat(self):
        print('Eating')

class Dog(Animal):
    def __init__(self, breed=None, role=None, dog_name=None, **kwargs):
        super().__init__(limbs=4, eyes=2, is_pet=True)
        self.breed = breed
        self.role = role
        self.deg_name = dog_name
        self.pees_on_tree = True
        #print(self, breed, role, name)
    
    def describe(self):
        s = f'''
        Animal: {type(self)},
        Breed: {self.breed},
        Role: {self.role},
        Name: {self.name},
        Pees on three: {self.pees_on_tree}
        '''
        print(s)
    
    def bark(self):
        print('Bark Bark Bark!')

        
class Spider(Animal):
    def __init__(self, poisonous=True, spider_name=None, **kwargs):
        super().__init__(limbs=8, eyes=5, is_pet=False)
        self.poisonous = poisonous
        self.spider_name = spider_name
        #print('from spider', 
        #      self, 
        #      'pois', poisonous, 
        #      'name', name)
        
    def make_net(self):
        print('Spider made a net!')
        
    def eat_fly(self):
        print('Eating fly, nam nam nam.')
        
    
class Chimera(Dog, Spider):
    def __init__(self, name):
        super().__init__(breed=r'Terrier', role='Rat Hunter', 
                         dog_name=name, spider_name=name, poisonous=True)
        
    def make_trick(self):
        self.make_net()
        self.eat_fly()
        self.bark()
```

```python
dog_spider = Chimera('Dog_spider')
```

```python
dog_spider.make_trick()
```

```python
class Rectangle:
    def __init__(self, length, height):
        self.length = length
        self.height = height
        
    def perimeter(self):
        return (self.length + self.height) * 2
    
    def area(self):
        return self.length * self.height
```

```python
r0 = Rectangle(3, 4)
r1 = Rectangle(5, 4)
r2 = Rectangle(9, 4)
r3 = Rectangle(2, 4)
```

```python
r2.area()
```

```python
class Square(Rectangle):
    def __init__(self, side):
        Rectangle.__init__(self, side, side)
        super().__init__(length=side, height=side)
```

```python
s = Square(6)
```

```python
s.perimeter()
```
