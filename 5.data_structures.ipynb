ლისტიდან წევრების ამოღება შესაძლებელია ინდექსების მიხედვით, მაგალითად:

```python
# ლისტის ტიპის კოლექციის გამოცხადება შერეული ტიპის წევრებით
items = [12, "Hello, World!", 3.14, [1, 2, 3], 1]
# len ფუნქცია დაგვიბრუნებს კოლექციის სიგრძეს
list_length = len(items)
# for ციკლი, რომელიც ციკლის ყოველ საფეხურზე დააბრუნებს 
# ინდექსებს ნულიდან კოლექციის ზომამდე
for i in range(list_length):
    # format მეტოდით ვაფორმატებთ სტრინგს, ხოლო items[i] დაგვიბრუნებს 
    # კოლექციის შესაბამის ინდექსზე ჩაწერილი წევრს
    print("index: {}, item: {}".format(i, items[i]))
```

```python
# ლისტის ტიპის კოლექციის გამოცხადება შერეული ტიპის წევრებით
items = [12, "Hello, World!", 3.14, [1, 2, 3], 1]
# len ფუნქცია დაგვიბრუნებს კოლექციის სიგრძეს
list_length = len(items)
# for ციკლი, რომელიც ციკლის ყოველ საფეხურზე დააბრუნებს 
# ინდექსებს ნულიდან კოლექციის ზომამდე
for i in range(list_length):
    # format მეტოდით ვაფორმატებთ სტრინგს, ხოლო items[i] დაგვიბრუნებს 
    # კოლექციის შესაბამის ინდექსზე ჩაწერილი წევრს
    if type(items[i]) == list:
        for item in items[i]:
            print(item)
    else:
        print("index: {}, item: {}".format(i, items[i]))
```

### len ფუნქცია
len ფუნქციას გადაეცემა ნებისმიერი (builtin) კოლექცია და გვიბრუნებს მის სიგრძეს, მაგ:

```python
c = [2, 3, 19, 'de', 2.34]
l = len(c)
print(l)
```

### append
append მეთოდით შესაძლებელია, ლისტისთვის წევრების დამატება თანმიმდევრობით, მაგალითად:

list ფუნქციით ნებისმიერი იტერირებადი ობიექტი გარდაიქმნება ლისტის ტიპის კოლექციად, მაგალითად:

```python
# range ფუნქცია გვიბრუნებს იტერირებად ობიექტს
rg = range(0, 20, 2)
print(rg)
# იტერირებადი ობიექტის გარდაქმნა ლისტის ტიპის კოლექციად
l = list(rg)
print(l)
```

### insert 
insert მეთოდით ლისტში შესაძლებელია წევრის ჩამატება ნებისმიერ ინდექსზე.

```python
a = [0, 1, 2, 3, 4]
# 123-ის ნულოვან ინდექსზე ჩამატება
a.insert(0, 123)
print(a)
# 45-ის მე-2 ინდექსზე ჩამატება
a.insert(2, 45)
print(a)
a.insert(-1, 99)
print(a)
```

### list concatenation

```python
s = "hello " + "world"
print(s)
```

### extend

### sort 
sort მეთოდი დაასორტირებს არსებულ კოლექციას.

### pop

# Tuple
Tuple-ის ტიპის კოლექცია შეგვიძლია გამოვაცხადოთ ნებისმიერი ტიპის წევრებით, ისევე როგორც ლისტის ტიპის კოლექცია, იმ განსხვავებით, რომ Tuple-ს ვაცხადებთ ერთხელ და შემდგომ მისი შეცვლა აღარ შეგვიძლია. Tuple-ზე არ შეგვიძლია არც წევრის წაშლა არც დამატება და არც შეცვლა.
ფუნქციებიდან აქვს მხოლოდ ორი ფუნქცია: count და index, რომლებიც ისევე მუშაობენ როგორც ლისტის ფუნქციები.

```python
# Tuple-ის გამოცხადება 
t = (1, 4, 4, 4, "asdf", 3.4, [1,2,3])
```
