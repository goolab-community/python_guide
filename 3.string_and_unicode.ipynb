## String formating % and format function

```python
#  αƒαƒ αƒαƒ›αƒαƒΆαƒαƒ£αƒαƒ αƒ¤αƒαƒ αƒ›αƒαƒΆαƒαƒ αƒ”αƒ‘αƒ
a_list = [1,2,3]
name = 'asdf'
float_number = 3.14
i = 34

print('float_number', float_number, 'name:', name, 'a_list:', a_list, 'i:', i)
```

### Formating with %

```python
print('float_number: %f, name: %s a_list: %s, i:%d'% (float_number, name, a_list, i))
```

### Formating with format function

```python
s = 'float_number: {}, name: {} a_list: {}, i:{}'.format(float_number, name, a_list, i)
```

```python
print(s)
```
