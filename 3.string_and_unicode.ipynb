## String formating % and format function

```python
#  პრიმიტიული ფორმატირება
a_list = [1,2,3]
name = 'asdf'
float_number = 3.14
i = 34

print('float_number', float_number, 'name:', name, 'a_list:', a_list, 'i:', i)
```

### Formating with %

```python
print('float_number: %f, name: %s a_list: %s, i:%d'% (float_number, name, a_list, i))
```

### Formating with format function

```python
s = 'float_number: {}, name: {} a_list: {}, i:{}'.format(float_number, name, a_list, i)
```

```python
print(s)
```

## UTF-8 (8-bit Unicode Transformation Format) 

|Number of bytes|Bits for code point|First code point|Last code point|Byte 1   |Byte 2  |Byte 3  |Byte 4 |    
| --------------:| ------------------:|---------------:|--------------:|--------:|-------:|-------:|------:|
|       1       |	      7	        |  U+0000	    |     U+007F	|0xxxxxxx |        |        |       |		
|       2       |	      11        |  U+0080	    |     U+07FF    |110xxxxx |10xxxxxx|	    |	    |
|       3       |	      16        |  U+0800	    |     U+FFFF	|1110xxxx |110xxxxx|10xxxxxx|       |

hello მოცემული ცხრილიდან კარგად ჩანს უნიკოდის ფორმატირების მოთოდი, რომლის მიხედვითაც, იმ შემთხვევაში თუ ჩარაქტერის ზომა 127 ბიტზე მცირეა, გამოიყენება მხოლოდ ერთი ბაიტი, ხოლო ბაიტში მაღალი თანრიგის ბიტში 0 (0xxxxxxx) მიუთითებს რომ ჩარაქტერისთვის გამოყოფილია მხოლოდ 127 ბიტი. თუ ბაიტის მაღალი თანრიგის პირველ ორ ბიტში მოცემულია 1 და 1 (110xxxxx), ეს მიუთითებს რომ ჩარაქტერი კოდირებულია ორ ბაიტში და ა.შ. 
UTF-8 ფორმატის გარდა, არსებობს მრავალი სხვა ფორმატი რომლებიც მორგებულია ამა თუ იმ პლათფორმას თუ ტექნოლოგიას. მაგალითად ზოგიერთი ვერსიის ვინდოუსის ოპერაციული სისტემის გარემოში, სტანსარტულად გამოიყენება CP1252 კოდირება, რომელიც გაკვეულწილად გასხვავდება UTF-8 კოდირებისგან. აუცილებელია დავიმახსოვროთ! დეკოდირებისთვის გამოვიყენოთ ის ფორმატი რა ფრომატშიც მოხდა ტექსტის კოდირება, წინააღმდეგ შემთხვევაში ყოველთვისაა იმის შესაძლებლობა, რომ დაიკარგოს ან დამახინჯდეს გარკვეული ინფორმაცია.
